```python
import requests

# Définir les paramètres de la requête
url = "https://newsapi.org/v2/everything"
params = {
    "q": "agriculture biologique OR bio",
    "language": "fr",
    "from": "2024-17-06",  # Date de début ajustée
    "to": "2024-17-07",    # Aujourd'hui
    "sortBy": "popularity",
    "apiKey": API_KEY
}


# Envoyer la requête
response = requests.get(url, params=params)

# Vérifier le succès de la requête
if response.status_code == 200:
    data = response.json()  # Convertir la réponse en JSON
    articles = data.get("articles", [])
    
    # Parcourir les articles pour afficher les informations principales
    for article in articles:
        print(f"Titre: {article['title']}")
        print(f"Auteur: {article['author']}")
        print(f"Source: {article['source']['name']}")
        print(f"Date: {article['publishedAt']}")
        print(f"Lien: {article['url']}\n")
else:
    print(f"Erreur: {response.status_code}, {response.text}")
```

```python
# Comptabiliser les articles par source
sources = [article["source"]["name"] for article in articles]
source_counts = Counter(sources)

# Afficher les résultats
print("Sources les plus actives :")
for source, count in source_counts.most_common():
    print(f"{source}: {count} articles")
```
