```python
# Définir les paramètres de la requête
url = "https://newsapi.org/v2/everything"
params = {
    "q": "bio",
    "language": "fr",
    "from": "2024-17-06",  # Date de début ajustée
    "to": "2024-17-07",    # Aujourd'hui
    "sortBy": "popularity",
    "apiKey": API_KEY
}


# Envoyer la requête
response = requests.get(url, params=params)

# Vérifier le succès de la requête
if response.status_code == 200:
    data = response.json()  # Convertir la réponse en JSON
    articles = data.get("articles", [])
    
    # Parcourir les articles pour afficher les informations principales
    for article in articles:
        print(f"Titre: {article['title']}")
        print(f"Auteur: {article['author']}")
        print(f"Source: {article['source']['name']}")
        print(f"Date: {article['publishedAt']}")
        print(f"Lien: {article['url']}\n")
else:
    print(f"Erreur: {response.status_code}, {response.text}")
```

```python
# Liste de mots-clés liés à l'agriculture biologique
keywords = ["agriculture", "production", "OGM", "cultures", "pesticide", "engrais", "agriculteur"]

# Fonction pour vérifier si un article est pertinent
def is_relevant(article, keywords):
    # Concaténer toutes les valeurs textuelles de l'article
    full_text = " ".join(
        str(value).lower() for value in article.values() if isinstance(value, str)
    )
    
    # Vérifiez si un mot-clé est présent dans le texte complet
    for keyword in keywords:
        if keyword in full_text:
            return True
    return False

# Filtrer les articles pertinents
relevant_articles = [article for article in articles if is_relevant(article, keywords)]

# Afficher les articles pertinents
print("Articles pertinents :")
for article in relevant_articles:
    print(f"- {article['title']}")
```

```python
from collections import Counter

# Comptabiliser les articles par source
sources = [article["source"]["name"] for article in articles]
source_counts = Counter(sources)

# Afficher les résultats
print("Sources les plus actives :")
for source, count in source_counts.most_common():
    print(f"{source}: {count} articles")
```
